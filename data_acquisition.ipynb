```python
import kagglehub
import shutil
import os
import numpy as np

random_seed = 42
np.random.seed(random_seed)
```

```python
download_path = kagglehub.dataset_download("codewithsk/tomato-leaf-disease-from-plantcity-2025")
dst = os.getcwd().replace('\\', '/') + '/data/raw/'

os.makedirs(dst, exist_ok=True)

for item in os.listdir(download_path):
    shutil.move(os.path.join(download_path, item), os.path.join(dst, item))
```

```python
# move all files into a single directory while avoiding overwrites

train_dir = "./data/raw/train/train"
test_dir = "./data/raw/test/test"
output_dir = "./data/raw/all_data"

os.makedirs(output_dir, exist_ok=True)

def move_files(src_dir):
    for class_name in os.listdir(src_dir):
        class_src = os.path.join(src_dir, class_name)
        if not os.path.isdir(class_src):
            continue

        class_dest = os.path.join(output_dir, class_name)
        os.makedirs(class_dest, exist_ok=True)

        for filename in os.listdir(class_src):
            src_file = os.path.join(class_src, filename)
            file_base, file_ext = os.path.splitext(filename)

            dest_file = os.path.join(class_dest, filename)

            # If filename exists, generate a new one
            counter = 1
            while os.path.exists(dest_file):
                new_filename = f"{file_base}_{counter}{file_ext}"
                dest_file = os.path.join(class_dest, new_filename)
                counter += 1

            shutil.move(src_file, dest_file)
            print(f"Moved: {src_file} → {dest_file}")

move_files(train_dir)
move_files(test_dir)

print("\n✅ Finished merging into all_data/ preserving class subfolders and avoiding name collisions.")
```

the folder <code>/data/raw/train</code>
 and <code>/data/raw/test</code> are from the original kaggle dataset structure and can be deleted after merging.
